### Индексация .loc и .iloc на примерах

```python
#Создадим серию с метками индекса и попробуем разные выборки по меткам и по позиции
s = pd.Series(list("abcdef"), index=[49, 48, 47, 0, 1, 2])
print(s)
```

```python
s.loc[0]    # выбирает данные по значению метки индекса 0: ‘d’.
```

```python
s.iloc[0]   # выбирает данные по позиции индекса 0: ‘a’.
```

```python
s.loc[0:1]  # выбирает данные по значению метки индекса от 0 до 1 (инклюзивно).
```

```python
s.iloc[0:1] # выбирает данные по позиции индекса от 0 до 1 (эксклюзивно).
```

```python
s.index
```

```python
# Бывает, что у датафрейма нет строки заголовков.
# Тогда для обращения к столбцам можно использовать метод loc / iloc.
# Импортируем данные и...
df_noheader = pd.read_csv('/content/mydata/icecream_data_CP1251.csv', header=None)
df_noheader = df_noheader.loc[1:] # ...и оставим данные без заголовка
df_noheader = df_noheader.reset_index(drop=True) # сбросим индекс строк
df_noheader.head()
```

```python
#выбрать значения на пересечении строк с индексом 0...1 и столбцов с индексом 2...3.
df_noheader.loc[0:1, 2:3]
```

```python
# Та же выборка с помощью iloc:
# выбрать значения на пересечении строк с индексом 0...2 (2 не включительно)
# и столбцов с индексом 2...4 (4 не включительно).
df_noheader.iloc[0:2, 2:4]
```

```python
#Поэкспериментируй с индексацией, убирая символ комментария и выполняя команды:

#df.loc[:, 'Флаеры'] #выбрать столбец по имени. Результат: Series
#df.loc[:, ['Флаеры']] #Результат: DataFrame
#df.iloc[:, 5] #выбрать столбец с меткой индекса 5 ('Флаеры'). Результат: Series
df.iloc[0:4, [5]] ##Результат: DataFrame
```

```python
# проверить тип возвращаемого объекта:
type(df.iloc[0:4, [5]]) # DataFrame
```
