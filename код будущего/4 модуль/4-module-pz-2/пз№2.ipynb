```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({'X': [18.2, 23.5, 108.4, 33.1, 44.9, 68.15, 20.8, 11.2, 22.4, 25.2, 60.3, 42.6],
                   'Y': [10.34, 12.32, 105.3, 23.9, 41.8, 71.5, 12.1, 0.95, 14.9, 14.9, 56.4, 33.05]})
df.head(12)
```

```python
df.plot.scatter(x = 'X', y = 'Y')
```

```python
df['X**2'] = df.X**2
df['Y**2'] = df.Y**2
df['X*Y'] = df.X*df.Y
df.head()
```

```python
A = np.array([[df['X**2'].sum(), df.X.sum()],
              [df.X.sum(), df.shape[0]]])
print(A)
```

```python
b = np.array([df['X*Y'].sum(), df['Y'].sum()])
print(b)
```

```python
k,b = list(np.linalg.solve(A, b))
```

```python
print(f'k = {k}, b = {b}')
```
