```python
# Выведи верхние строки датасета
titanic.head(10)
```

```python
# Выведи описательные статистики
titanic.describe()
```

```python
# Выведи максимальное значение признака ‘Fare’.
titanic.Fare.max()
```

```python
# Получи срез [2:4,'Name']
titanic.loc[2:4, 'Name']
```

```python
# Выведи на печать число женщин и мужчин
print(f"Женщин: {len(titanic[titanic.Sex == 'female'])}")
print(f"Мужчин: {len(titanic[titanic.Sex == 'male'])}")
```

```python
# Выведи число выживших пассажиров 1 класса. Используй логический оператор '&'.
print(len(titanic[(titanic.Pclass == 1) & (titanic.Survived == 1)]))
```

```python
# Выведи средний возраст выживших с использованием .loc()
titanic.loc[titanic.Survived == 1, 'Age'].mean()
```

```python
# Выведи средний возраст выживших иным способом
titanic[titanic.Survived == 1].Age.mean()
```
