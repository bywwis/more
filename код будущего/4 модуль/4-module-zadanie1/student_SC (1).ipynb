```python
# Размести на Google Drive файл students-mat.csv из материалов к заданию и загрузи его в dataframe
# Описание датасета: https://www.kaggle.com/uciml/student-alcohol-consumption

student = pd.read_csv('/content/drive/MyDrive/mydata/student-mat.csv')
```

```python
# Выведи верхние строки датасета
student.head()
```

```python
# Выведи описательные статистики
student.describe()
```

```python
# Оставь в датафрейме только признаки из этого списка:
cols = ['sex', 'age', 'romantic', 'freetime', 'studytime', 'paid', 'goout', 'G3']
student = student[cols]
print(student)
```

```python
#Средняя итоговая оценка у тех, кто берет дополнительные платные занятия, и у тех, кто нет.
print(student[student.paid == 'yes'].G3.mean())
print(student[student.paid == 'no'].G3.mean())
```

```python
# Получи срез [3:6,'G3']
student.loc[3:6, 'G3']
```

```python
# Выведи на печать число женщин и мужчин
print(f"Женщин: {len(student[student.sex == 'F'])}")
print(f"Мужчин: {len(student[student.sex == 'M'])}")
```

```python
# Выведи число мужчин, имеющих романтические отношения. Используй логический оператор '&'.
len(student[(student.sex == 'M') & (student.romantic == 'yes')])
```

```python
# Выведи медианное свободное время после школы у женщин
# Мы не можем брать среднее потому что эта переменная - не непрерывная, а порядковая категориальная переменная.
student[student.sex == 'F']['freetime'].median()
```

```python
# Средний возраст женщин .loc()
student.loc[student.sex == 'F', 'age'].mean()
```

```python
# Средний возраст мужчин, по-другому
student[student.sex == 'M'].age.mean()
```
